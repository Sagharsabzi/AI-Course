# Simple Linear Regression

In this notebook, we introduce **Simple Linear Regression**.

Goals:
- Understand the relationship between two numerical variables
- Learn how a straight line can model this relationship

The linear regression model is defined as:

y = w * x + b

Where:
- w is the slope
- b is the intercept

```python
import numpy as np
import matplotlib.pyplot as plt
```

We use a simple dataset where:
- x represents the area of a house (in square meters)
- y represents the house price

```python
X = np.array([60, 75, 80, 90, 100, 110, 120, 130, 140, 160])
y = np.array([180, 220, 260, 300, 340, 380, 420, 460, 500, 580])
```

First, we visualize the data to see if a linear relationship exists.

```python
plt.scatter(X, y)
plt.xlabel("Area (mÂ²)")
plt.ylabel("Price")
plt.title("House Price vs Area")
plt.show()
```

Before using any machine learning library, we manually define a linear model.

This helps us understand:
- The effect of the slope (w)
- The effect of the intercept (b)

```python
w = 3.0
b = 10.0

y_pred = w * X + b
```

```python
plt.scatter(X, y, label="Actual Data")
plt.plot(X, y_pred, color="red", label="Predicted Line")
plt.xlabel("Area (mÂ²)")
plt.ylabel("Price")
plt.title("Simple Linear Regression (Manual)")
plt.legend()
plt.show()
```

The model is not perfect.

For each data point, the difference between the real value and the predicted value is called **error**:

error = y_actual â y_predicted

```python
errors = y - y_pred
mse = np.mean(errors ** 2)

mse
```
